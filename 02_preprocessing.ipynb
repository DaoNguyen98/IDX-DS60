## Load and Combine Data

```python
import glob
import os

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, StandardScaler

pd.set_option("display.max_columns", 100)
sns.set_theme(style="whitegrid")

print("Imports successful")
```

```python
# Point this at the local folder where you downloaded the CRMLSSold files via FileZilla
DATA_DIR = "C:/Users/thanh/Downloads"  # <-- update to your local path

files = sorted(glob.glob(os.path.join(DATA_DIR, "CRMLSSold*")))
print(f"Found {len(files)} files:")
for f in files:
    print(" -", os.path.basename(f))

assert len(files) >= 6, "Task requires a minimum of 6 months of data — check DATA_DIR."
```

```python
dfs = []
for f in files:
    tmp = pd.read_csv(f, low_memory=False)
    tmp["__source_file"] = os.path.basename(f)
    dfs.append(tmp)

df_raw = pd.concat(dfs, ignore_index=True)
print(df_raw.shape)
df_raw.head()
```

```python
df = df_raw[
    (df_raw["PropertyType"] == "Residential")
    & (df_raw["PropertySubType"] == "SingleFamilyResidence")
].copy()

print(f"Rows before filter: {len(df_raw):,}")
print(f"Rows after filter:  {len(df):,}")
```

## Handle Missing Values

```python
missing = df.isnull().sum().sort_values(ascending=False)
missing.head(20)
```

```python
numeric_cols = df.select_dtypes(include=["number"]).columns

for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
# Drop columns that are entirely missing
df = df.dropna(axis=1, how="all")
```

```python
categorical_cols = df.select_dtypes(include=["object", "string"]).columns

for col in categorical_cols:
    if df[col].mode().empty:
        print(f"{col} has no mode (all values are missing)")
    else:
        df[col] = df[col].fillna(df[col].mode()[0])
```

## Encode Categorical Variables

```python
encoder = LabelEncoder()

for col in categorical_cols:
    df[col] = encoder.fit_transform(df[col].astype(str))
```

## Scale Numerical Features

```python
scaler = StandardScaler()

numeric_features = df.select_dtypes(include=["number"]).columns.drop("ClosePrice")

df[numeric_features] = scaler.fit_transform(df[numeric_features])
```

## Create Train/Test Split

```python
df["CloseDate"] = pd.to_datetime(df["CloseDate"])

latest_month = df["CloseDate"].dt.to_period("M").max()

test_df = df[df["CloseDate"].dt.to_period("M") == latest_month]

train_df = df[df["CloseDate"].dt.to_period("M") < latest_month]
```

```python
print("Training shape:", train_df.shape)
print("Testing shape:", test_df.shape)

print(train_df["CloseDate"].min(), "-", train_df["CloseDate"].max())
print(test_df["CloseDate"].min(), "-", test_df["CloseDate"].max())
```

```python
import os

os.makedirs("processed", exist_ok=True)

train_df.to_csv("processed/train_cleaned.csv", index=False)
test_df.to_csv("processed/test_cleaned.csv", index=False)

print("Files saved successfully.")
```
